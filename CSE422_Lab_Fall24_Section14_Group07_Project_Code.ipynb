```python
def display_occurrence_counts(occurrence_series, target_value):
    print(f"=== Occurrences of {target_value} per Column ===")
    for column, count in occurrence_series.items():
        print(f"{column}: {count}")
    print("=========================================\n")
```

```python
def count_specific_value_occurrences(df, target_value):
    occurrence_counts = df.eq(target_value).sum()
    return occurrence_counts
```

```python
if air_quality_df is not None:
    # Specify the target value to count
    target_value = -200

    # Count occurrences of the target value per column
    negative_200_counts = count_specific_value_occurrences(air_quality_df, target_value)

    # Display the occurrence counts
    display_occurrence_counts(negative_200_counts, target_value)
```

```python
def replace_negative_values(df, target_value=-200, replacement=np.nan):
    if df is not None:
        modified_df = df.copy()
        occurrences_before = (modified_df == target_value).sum().sum()
        modified_df.replace(to_replace=target_value, value=replacement, inplace=True)
        occurrences_after = (modified_df == target_value).sum().sum()
        print(f"Replaced {occurrences_before} instances of {target_value} with {replacement}.\n")
        return modified_df
    else:
        print("Error: DataFrame is None. Cannot perform replacement.")
        return df
```

```python
def select_columns_for_mean(df, start_idx=2, end_idx=14):
    if df is not None:
        selected_columns = df.iloc[:, start_idx:end_idx]
        print(f"Selected columns from index {start_idx} to {end_idx - 1} for mean calculation.\n")
        return selected_columns
    else:
        print("Error: DataFrame is None. Cannot select columns.")
        return df
def compute_column_means(df_selected):
    if df_selected is not None:
        column_means = df_selected.mean()
        return column_means
    else:
        print("Error: Selected DataFrame is None. Cannot compute means.\n")
        return None

def display_column_means(means_series):
    if means_series is not None:
        print("=== Mean Values per Selected Column ===")
        for column, mean in means_series.items():
            print(f"{column}: {mean:}")
        print("=======================================\n")
    else:
        print("Error: Means series is None. Cannot display means.\n")
```

```python
def fill_missing_with_mean(df, start_idx=2, end_idx=14):
    if df is not None:
        selected_columns = df.iloc[:, start_idx:end_idx]
        mean_values = selected_columns.mean()
        df_filled = df.copy()
        df_filled.iloc[:, start_idx:end_idx] = selected_columns.fillna(mean_values)
        print(f"Filled NaN values in columns {start_idx} to {end_idx - 1} with their respective means.\n")
        return df_filled
    else:
        print("Error: DataFrame is None. Cannot fill missing values.\n")
        return df
```

```python
def check_data_types(df):
    if df is not None:
        dtypes = df.dtypes
        print("=== Data Types of Each Column ===")
        for column, dtype in dtypes.items():
            print(f"{column}: {dtype}")
        print("=================================\n")
        return dtypes
    else:
        print("Error: DataFrame is None. Cannot check data types.\n")
        return None
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
def encode_categorical_variables(df):
    if df is not None:
        encoder = LabelEncoder()
        categorical_columns = df.select_dtypes(include=["object"]).columns
        if len(categorical_columns) == 0:
            print("No categorical columns found to encode.\n")
            return df
        print("=== Encoding Categorical Variables ===")
        for col in categorical_columns:
            try:
                # Ensure the column is treated as a 1D array
                df[col] = encoder.fit_transform(df[col])
                print(f"Encoded column '{col}' successfully.")
            except Exception as e:
                print(f"Error encoding column '{col}': {e}")
        print("======================================\n")
        return df
    else:
        print("Error: DataFrame is None. Cannot encode categorical variables.\n")
        return df
```

```python
def show_distributions(df, categories, figsize=(25, 5), color="darkseagreen", kde=True):
    if df is not None:
        # Validate categories
        valid_categories = [col for col in categories if col in df.columns]
        invalid_categories = set(categories) - set(valid_categories)
        if invalid_categories:
            print(f"Warning: The following categories are not in the DataFrame and will be skipped: {invalid_categories}\n")

        num_plots = len(valid_categories)
        if num_plots == 0:
            print("No valid categories provided for distribution plots.\n")
            return

        # Determine subplot grid size
        cols = min(num_plots, 3)  # Up to 3 columns per row
        rows = (num_plots + cols - 1) // cols  # Ceiling division for rows

        fig, axes = plt.subplots(rows, cols, figsize=figsize)
        axes = axes.flatten()  # Flatten in case of multiple rows

        for idx, category in enumerate(valid_categories):
            sns.histplot(data=df, x=category, kde=kde, color=color, ax=axes[idx])
            axes[idx].set_title(f"Distribution of {category}", fontsize=16)
            axes[idx].set_xlabel(category, fontsize=14)
            axes[idx].set_ylabel("Frequency", fontsize=14)

        # Hide any unused subplots
        for idx in range(len(valid_categories), len(axes)):
            fig.delaxes(axes[idx])

        plt.tight_layout()
        plt.show()
        print(f"Displayed distributions for categories: {valid_categories}\n")
    else:
        print("Error: DataFrame is None. Cannot display distributions.\n")
```

```python
def create_scatter_plot(dataframe, x_axis, y_axis, color_group, plot_title):
    scatter_fig = px.scatter(
        dataframe,
        x=x_axis,
        y=y_axis,
        color=color_group,
        title=plot_title
    )
    scatter_fig.show()

# Filled Area Plot Visualization
def create_area_plot(dataframe, x_axis, y_axis, color_group, hover_info):
    area_fig = px.area(
        dataframe,
        x=x_axis,
        y=y_axis,
        color=color_group,
        hover_data=[hover_info]
    )
    area_fig.show()

# Line Charts Visualization
def create_line_charts(dataframe, y_columns, x_axis, chart_title):
    line_fig = go.Figure()
    for column in y_columns:
        line_fig.add_trace(go.Scatter(
            x=dataframe[x_axis],
            y=dataframe[column],
            mode='lines',
            name=column
        ))
    line_fig.update_layout(title=go.layout.Title(text=chart_title))
    line_fig.show()

# Violin Plots Visualization
def plot_violin_charts(x1, y1, x2, y2, x3, y3, dataframe):
    fig, axes = plt.subplots(1, 3, figsize=(25, 5))

    sns.violinplot(x=x1, y=y1, data=dataframe, hue=x1, palette='rocket', ax=axes[0])
    axes[0].set_title(f"{x1} by {y1}")

    sns.violinplot(x=x2, y=y2, data=dataframe, hue=x2, palette='rocket', ax=axes[1])
    axes[1].set_title(f"{x2} by {y2}")

    sns.violinplot(x=x3, y=y3, data=dataframe, hue=x3, palette='rocket', ax=axes[2])
    axes[2].set_title(f"{x3} by {y3}")

    plt.tight_layout()
    plt.show()

# Heatmap Visualization
def generate_heatmap(width, height, heatmap_title, dataframe):
    plt.figure(figsize=(width, height))
    plt.title(heatmap_title)
    sns.heatmap(dataframe, annot=True, cmap='flare')
    plt.show()
```

```python
def render_correlation_heatmap(width, height, heatmap_title, correlation_data):
    # Initialize the plotting area with specified dimensions
    heatmap_fig = plt.figure(figsize=(width, height))

    # Set the title of the heatmap
    plt.title(heatmap_title, fontsize=16)

    # Create the heatmap using seaborn
    sns.heatmap(
        correlation_data,
        annot=True,
        cmap='flare',
        fmt=".2f",  # Format annotations to two decimal places
        linewidths=0.5,  # Width of the lines that will divide each cell
        linecolor='white',  # Color of the lines
        cbar_kws={'shrink': 0.8}  # Shrink the color bar for better fit
    )

    # Adjust layout for better spacing
    plt.tight_layout()

    # Display the heatmap
    plt.show()
```

```python
# Define the classifier names
classifiers = ["Decision Tree Classifier", "Random Forest Classifier", "K-Nearest Neighbor"]

# Define the performance metrics
metrics = {
    'Recall': model_train_recall_scores,
    'Precision': model_train_precision_scores,
}

# Set up the bar chart parameters
num_classifiers = len(classifiers)
num_metrics = len(metrics)
bar_width = 0.3
indices = np.arange(num_classifiers)

# Initialize the plot
fig, ax = plt.subplots(figsize=(10, 7), constrained_layout=True)

# Plot each metric as a separate set of bars
for idx, (metric_name, scores) in enumerate(metrics.items()):
    # Calculate the position for each set of bars
    bar_positions = indices + idx * bar_width
    # Create the bars
    bars = ax.bar(bar_positions, scores, width=bar_width, label=metric_name)
    # Add labels to the bars
    ax.bar_label(bars, labels=[f"{score:.2f}" for score in scores], padding=5)

# Customize the x-axis
mid_point = bar_width * (num_metrics - 1) / 2
ax.set_xticks(indices + mid_point)
ax.set_xticklabels(classifiers)

# Add title and labels
ax.set_title('Training Phase: Recall and Precision Metrics')
ax.set_ylabel('Scores')

# Add a legend
ax.legend(title='Metrics', loc='upper left', ncol=num_metrics)

# Set y-axis limits
ax.set_ylim(0, 1.5)

# Display the plot
plt.show()
```
